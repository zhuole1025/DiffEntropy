```python
import torch
import os

input_dir = '/home/pgao/zl/zl/DiffEntropy/Next-DiT-ImageNet/results'
all_files = os.listdir(input_dir)
all_files = [file_i for file_i in all_files if '.pt' in file_i]
all_files = sorted(all_files, key=lambda x: float(x.split('.pt')[0].split('_')[-1]))
data = []
for file_i in all_files:
    data.append(torch.load(os.path.join(input_dir, file_i)).mean(dim=0)) 

```

```python
import matplotlib.pyplot as plt

# draw a line for each data
plt.figure(figsize=(10, 5))
for i in range(0, len(data), 4):
    plt.plot(data[i])

plt.show()
```
